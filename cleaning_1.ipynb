```python
from collections import Counter
```

```python
## Problem ID count
c = Counter(dat['problemId'])
c.most_common()
```

```python
## IDCount count
d = Counter([i[1] for i in list(c.items())])
d.most_common()
list(d.values())
```

```python
## # of problem that have IDcount greater than 30
sum = 0
for i in d.items():
    if i[0]> 30:
        sum+=i[1]
sum
```

```python
stu_count = Counter()
for pid in dat['problemId'].unique():
    stu_count[pid] = len(dat[dat['problemId']==pid]['studentId'].unique())
```

```python
stu_count
```

```python
## stu_count count
scc = Counter([i[1] for i in list(stu_count.items())])
scc.most_common()
```

```python
## # of problem that have IDcount greater than 30
sum = 0
for i in scc.items():
    if i[0]> 30:
        sum+=i[1]
sum
```
